```python
years = np.array(list(data.keys()))
```

```python
%matplotlib inline

plt.plot(years,people, label= r"$Spassk-Dalny$")
plt.grid(True)
plt.legend()
plt.show
```

```python
solve = odeint(dxdt, y0=1, t=years) 
solve = solve.flatten()
```

```python
%matplotlib inline

plt.plot(years,people, label= r"$Spassk-Dalny$")
plt.plot(years,solve, label= r"$x' = kx$")
plt.grid(True)
plt.legend()
plt.show()
```
